```python
import joblib
import numpy as np
```

```python
model = joblib.load("../model/saved_model.pkl")
```

```python
def suggest_career(skills, interests):
    user_input = skills.lower() + " " + interests.lower()
    
    probabilities = model.predict_proba([user_input])[0]
    careers = model.classes_
    
    top_indices = probabilities.argsort()[-3:][::-1]
    
    results = [(careers[i], round(probabilities[i]*100, 2)) for i in top_indices]
    
    return results
```

```python
suggest_career("python machine learning", "ai research")
```
